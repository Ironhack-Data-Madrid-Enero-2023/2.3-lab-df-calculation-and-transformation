## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

```python
# import required libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Next, import `Orders` from Ironhack's database into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
orders = pd.read_csv('orders.csv')
```

```python
orders.head()
```

```python
customer = orders.groupby(by='CustomerID').sum()

customer.head()
```

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

```python
percentil_95 = np.percentile(customer['amount_spent'],95)
percentil_75 = np.percentile(customer['amount_spent'],75)

customer['tipo_cliente'] = customer['amount_spent'].apply(lambda x: 'vip' if x > percentil_95 else 'preferred' if percentil_95 > x > percentil_75 else 'normal')
```

```python
customer['tipo_cliente'].unique()
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
orders['tipo_cliente'] = customer['tipo_cliente']
```

```python
merged = orders.merge(customer, on='tipo_cliente')

merged.head()
```

```python
merged.groupby(by='Country').tipo_cliente.value_counts()
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
percentil_75 = np.percentile(customer['amount_spent'],75)

customer['vip_pref'] = customer['amount_spent'].apply(lambda x: 'vip_pref' if x > percentil_75 else 'normal')
```

```python
orders['vip_pref'] = customer['vip_pref']
```

```python
merged2 = orders.merge(customer, on='vip_pref')
```

```python
merged2.groupby(by='Country').vip_pref.value_counts()
```
