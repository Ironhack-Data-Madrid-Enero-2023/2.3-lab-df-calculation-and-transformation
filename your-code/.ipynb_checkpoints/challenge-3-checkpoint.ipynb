## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

```python
#1 VIP Customers =  aggregated expenses > 95th quantile
#2 Preferred Customers = 75th quantile < aggregated expenses > 95thquantile 

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

Next, import `Orders` from Ironhack's database into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
orders = pd.read_csv('Orders.csv')
orders.head() #amount_spent
```

```python
def quant (x):
    return np.percentile(x, 95)
```

```python
Clients = orders.pivot_table(index=['CustomerID'],  aggfunc= sum)['amount_spent'].sort_values(ascending=False)
```

```python
quantile_range = orders.groupby('CustomerID')['amount_spent'].sum().quantile([0.75,0.95])
print(quantile_range) 
orders[orders.groupby('CustomerID')['amount_spent'].sum().between(quantile_range.iloc[0], quantile_range.iloc[1])]
quant95 = 5840.182
quant75 = 1661.640

#Hemos encontrado que el quartil 95 esta a partir de 5840.182
#Hemos encontrado que el quartil 75 esta a partir de 1661.640


```

```python
client= Clients.to_dict()
VIP = []
Pref = []
for k,v in client.items():
    if v > (5840.182):
        VIP.append(k)

for k,v in client.items():
    if v < (5840.182) and v > 1661.640:
        Pref.append(k)  
```

```python
Neww= []
Newww= []

for e in orders.CustomerID:
    if e in VIP:
        Neww.append('VIP')
    else : 
        Neww.append('None')
        
for e in orders.CustomerID:
    if e in Pref:
        Newww.append('Pref')
    else :
        Newww.append('None')

        
orders['Vips'] = Neww
orders['Prefs'] = Newww

orders.head(20)
```
